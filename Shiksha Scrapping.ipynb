```python
output = []

for item in range(len(placements_links_engineering)):
    response = requests.get(placements_links_engineering[item], headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    
    link = placements_links_engineering[item]
    
    name = clg_names[item]
    programm = "Engineering"
    try:
        review = soup.find("span", class_="rating-block rvw-lyr").text
    except:
        review = ""
    years = list()
    
    heads = soup.find_all('h2')
    #print(heads)
    headpresent = False
    try:
        for i in heads:
            #print(i.text)
            if "highlights" in i.text.strip().lower() or "batch placed" in i.text.strip().lower():
                headpresent = True
            
            if headpresent == True:
                if "highlights" in i.text.strip().lower() or "batch placed" in i.text.strip().lower():# or "average package" in i.text.strip().lower() or "highest package" in i.text.strip().lower():
                    print(i.text)
                    print("-----------")
                    table = i.find_next('table')
                    #print(table)
                    headings = table.find_all('th')
                    #print(headings)
                    for j in headings:
                        val = j.text.strip()
                        val = re.findall('\d\d\d\d', val)
                        #print(val)
                        if val == []:
                            continue  

                        years.append(val[0])
                    #print(years)

                    rows = table.find_all('tr')
                    #print(rows)
                    for yearcount in range(0, 1):

                        high_pkg = ""
                        avg_pkg = ""
                        med_pkg = ""
                        comp_part = ""
                        offer_made = ""
                        ppo = ""
                        inter_offer = ""
                        stud_reg = ""
                        stud_plac = ""
                        plac_per = ""

                        for i in rows:
                            td = i.text.strip().lower()
                            #print(td)

                            if "highest package (domestic)" in td or "btech highest package" in td or "highest domestic package" in td or "highest salary" in td or "highest package (engineering)" in td or "highest package" in td:
                                high_pkg = i.find_all('td')[yearcount + 1].text.strip()
                                #print("high_pkg: ",high_pkg)

                            if "btech cse average package" in td or "average package" in td or "average salary" in td or "domestic average package" in td:
                                avg_pkg = i.find_all('td')[yearcount + 1].text.strip()
                                #print("avg_pkg: ",avg_pkg)

                            if "median package" in td or "median salary" in td:
                                med_pkg = i.find_all('td')[yearcount + 1].text.strip()
                                #print("med_pkg: ",med_pkg)

                            if "companies participated" in td or "no. of companies registered" in td or "companies visited" in td or "total no. of companies visited" in td or "no. of recruiters" in td or "total recruiters" in td or "no. of companies" in td or "no. of companies visited" in td or "no. of companies participated" in td or "total number of companies registered" in td or "total companies participated" in td:
                                comp_part = i.find_all('td')[yearcount + 1].text.strip()
                                #print("comp_part: ",comp_part)

                            if "total offers" in td or "offers made" in td or "offers received" in td or "no. of offers received" in td or "offers made" in td or "no. ofjoboffers" in td or "total jobs offers" in td or "no. of offers made" in td or "no. of offers" in td or "total offers made" in td or "number of offers" in td:
                                offer_made = i.find_all('td')[yearcount + 1].text.strip()
                                #print("offer_made: ",offer_made)

                            if "ppos received" in td or "ppos offered" in td or "number of ppos" in td or "ppos" in td or "no. of ppos received" in td:
                                ppo = i.find_all('td')[yearcount + 1].text.strip()
                                #print("ppo: ",ppo)

                            if "international offers made" in td or "no. of international offers" in td or "international offers" in td:
                                inter_offer = i.find_all('td')[yearcount + 1].text.strip()
                                #print("inter_offer: ",inter_offer)

                            if "no. of students registered" in td or "total students" in td:
                                stud_reg = i.find_all('td')[yearcount + 1].text.strip()
                                #print("stud_reg: ",stud_reg)

                            if "total students placed" in td or "no. of students placed" in td or "no. of students received offers" in td or "students placed" in td:
                                stud_plac = i.find_all('td')[yearcount + 1].text.strip()
                                #print("stud_placed: ", stud_plac)

                            if "placement percentage" in td or "placement rate" in td:
                                plac_per = i.find_all('td')[yearcount + 1].text.strip()
                                #print("plac_per: ",plac_per)

                        print("Name: ", name)
                        if years == []:
                            print("Year: ", years)
                        else:
                            print("Year: ", years[0])
                        print("high_pkg: ",high_pkg)
                        print("avg_pkg: ",avg_pkg)
                        print("med_pkg: ",med_pkg)
                        print("comp_part: ",comp_part)
                        print("offer_made: ",offer_made)
                        print("ppo: ",ppo)
                        print("inter_offer: ",inter_offer)
                        print("stud_reg: ",stud_reg)
                        print("stud_placed: ", stud_plac)
                        print("plac_per: ",plac_per)

                        data = {
                            "Link": link,
                            "Name": name,
                            "Year": years,
                            "Review": review,
                            "Highest CTC": high_pkg,
                            "Average CTC": avg_pkg,
                            "Median CTC" : med_pkg,
                            "Companies Participated": comp_part,
                            "Offer Made": offer_made,
                            "Preplacement Offer": ppo,
                            "International Offer": inter_offer,
                            "Students Registered": stud_reg,
                            "Students Placed": stud_plac,
                            "Placement Percentage": plac_per
                             }
                        output.append(data)
                    break
                
        if headpresent == False:
            print("Link:", link)
            print("Name: ", name)
            print("Info Not available")
            
            data2 = {
                "Link": link,
                "Name": name,
                "Year": "Placement Data Not Available"
            }
            output.append(data2)
                    
    except:
        pass
    print("-------------------- ", placements_links_engineering[item])
```
